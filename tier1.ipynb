```python
import networkx as nx
import matplotlib.pyplot as plt
import random as random

random.seed(45)
G = nx.Graph()

districts = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
G.add_nodes_from(districts)

edges = [
    # Original edges
    ('A', 'B'), ('A', 'L'), 
    ('B', 'Y'),
    ('C', 'F'), ('C', 'M'),
    ('D', 'T'),
    ('E', 'K'), ('E', 'G'), ('E', 'N'), 
    ('G', 'J'), 
    ('H', 'J'),
    ('I', 'K'), ('I', 'O'),
    ('K', 'O'), 
    ('L', 'O'), ('L', 'M'), 
    ('M', 'U'),
    ('N', 'O'),

    # New connections (covering P–Z and improving connectivity)
    ('Q', 'R'),
    ('S', 'T'),
    ('A', 'V'), ('V', 'L'), ('W', 'X'),
    ('X', 'Y'), ('Y', 'Z'),

    # Cross-links (VERY IMPORTANT for path diversity)
    ('R', 'L'),
    ('T', 'H'), ('U', 'J'),
    ('S', 'D'), ('X', 'G'),
    ('Z', 'N'),

    # Extra shortcuts (to avoid linear chains)
    ('C', 'P'), ('F', 'Q'),
    ('K', 'S'),
    ('M', 'U'),
]

G.add_edges_from(edges)

plt.figure(figsize=(8,6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=2000, font_size=10)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()

```
